```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
# Load the dataset
dataset = pd.read_csv('/content/new_model.csv')
```

```python
#the target column is named 'Class'
X = dataset.drop('Class', axis=1)  # Features
y = dataset['Class']  # Target variable
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Create a KNN classifier with a specified number of neighbors (k)
k = 3
knn_classifier = KNeighborsClassifier(n_neighbors=k)
```

```python
# Fit the model to the training data
knn_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = knn_classifier.predict(X_test)
```

```python
# Compute the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Visualize the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
