```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
dataset = pd.read_csv('/content/new_model.csv')
```

```python
X = dataset.drop('Class', axis=1)  # Features
y = dataset['Class']  # Target variable
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Create lists to store accuracy values for different values of n_estimators
n_estimators_values = [10, 50, 100, 200]  # You can adjust the values as needed
accuracy_scores = []
```

```python
# Loop through different values of n_estimators and calculate accuracy
for n_estimators in n_estimators_values:
    rf_classifier = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    rf_classifier.fit(X_train, y_train)
    y_pred = rf_classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
```

```python
# Visualize accuracy results as a graph
plt.figure(figsize=(8, 6))
plt.plot(n_estimators_values, accuracy_scores, marker='o', linestyle='-')
plt.title('Random Forest Accuracy vs. Number of Estimators')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.xticks(n_estimators_values)
plt.grid(True)
plt.show()
```
